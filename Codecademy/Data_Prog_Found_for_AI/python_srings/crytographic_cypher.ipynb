```python
# Encoding function
def caesar_cipher_encode(text, shift):
    result = []
    for char in text:
        if char.isupper():
            result.append(chr((ord(char) + shift - 65) % 26 + 65))
        elif char.islower():
            result.append(chr((ord(char) + shift - 97) % 26 + 97))
        else:
            result.append(char)
    return ''.join(result)

# Decoding function
def caesar_cipher_decode(text, shift):
    result = []
    for char in text:
        if char.isupper():
            result.append(chr((ord(char) - shift - 65) % 26 + 65))
        elif char.islower():
            result.append(chr((ord(char) - shift - 97) % 26 + 97))
        else:
            result.append(char)
    return ''.join(result)

# Brute-force function to try all possible shifts
def bf_caesar_cipher(ciphertext):
    for shift in range(1, 26):
        result = ""
        for char in ciphertext:
            if char.isupper():
                result += chr((ord(char) - shift - 65) % 26 + 65)
            elif char.islower():
                result += chr((ord(char) - shift - 97) % 26 + 97)
            else:
                result += char
        print(f"Shift {shift}: {result}")

# Test the functions
plain_text = "this is a test of the system"
shift = 14


# Encoding
encoded_text = caesar_cipher_encode(plain_text, shift)
print("Encoded:", encoded_text)

# Decoding
decoded_text = caesar_cipher_decode(encoded_text, shift)
print("Decoded:", decoded_text)

# Brute-force a ciphertext without a known shift
ciphertext = ""
print("\nBrute-forcing the ciphertext:")
bf_caesar_cipher(ciphertext)
```
